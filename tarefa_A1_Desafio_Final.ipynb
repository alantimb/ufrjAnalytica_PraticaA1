```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("Dataset_Processo_Seletivo_UFRJ_Analytica_2022.1.csv")
```

```python
df_ano1991 = df[ df["ano"] == 1991 ]
df_ano2010 = df[ df["ano"] == 2010 ]
df_diferencaExpVidas =  abs(df_ano1991["expectativa_vida"].values - df_ano2010["expectativa_vida"].values)

x = df_ano1991["sigla_uf"].values # Siglas dos estados
y = df_diferencaExpVidas # Diferenças entre as expectativas de vida

df2 = pd.DataFrame(
    {
        "siglas_uf": x,
        "diferenca_expectativa_vida": y
    }
)

plt.bar(x, y, width=0.75, edgecolor='black')

plt.xticks(fontsize=8)
plt.yticks(np.arange(0, 14, 1), np.arange(0, 14, 1), fontsize=8)
plt.xlabel('Estados Brasileiros', fontsize=15)
plt.ylabel('Diferença de Expectiva de Vida', fontsize=15)
plt.title('Expectiva de Vida entre 1991 e 2010', fontsize=18)

plt.show()    
```

```python
estados_lista = []
# Separa os estados com expectativa de vida acima de 10 anos
for i in range(0, len(df2.diferenca_expectativa_vida.values)-1):
    if df2.diferenca_expectativa_vida.values[i] > 10:
        estados_lista.append(df2.siglas_uf.values[i])
# Converte esses estados em string
estados = ''
for i in range(0, len(estados_lista)):
    if i == len(estados_lista)-1:
        estados = estados + estados_lista[i]
        break        
    elif i == len(estados_lista)-2:
        estados = estados + estados_lista[i] + " e "
    else:
        estados = estados + estados_lista[i] + ", "

resposta = f'''Item 4. 
Os estados que tiveram um aumento de pelo menos 10 anos na expectativa de vida entre 1991 e 2010 foram: 
{estados}.
'''
print(resposta)
```
