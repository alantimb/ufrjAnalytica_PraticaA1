```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("Dataset_Processo_Seletivo_UFRJ_Analytica_2022.1.csv")
```

```python
df_anoFiltrado = df[ df["ano"] == 1991]
x = df_anoFiltrado.sigla_uf.values
y = df_anoFiltrado.populacao_urbana.values

plt.bar(x, y, width=0.8, edgecolor='black')

plt.xlabel('Estados Brasileiros', fontsize=15)
plt.ylabel('População Urbana', fontsize=15)
plt.title('Gráfico populacional brasileiro em 1991', fontsize=18)

plt.show()
```

```python
resposta = f'''Item 1.
A partir do gráfico é possível observar a alta discrepância entre a população urbana de alguns estados do Brasil.
Os que mais se destoam são os estados do norte e sudeste, os quais possuem, respectivamente, as menores e maiores \nproporções de população urbana do país.
'''
print(resposta)
```

```python
df_ano1991Filtrado = df[ df["ano"] == 1991]
df_ano2000Filtrado = df[ df["ano"] == 2000]
df_ano2010Filtrado = df[ df["ano"] == 2010]

x1 = df_ano1991Filtrado.idhm.values
x2 = df_ano2000Filtrado.idhm.values
x3 = df_ano2010Filtrado.idhm.values

plt.hist(x1, 10, rwidth=0.5, color='lightblue', edgecolor='black')
plt.hist(x2, 10, rwidth=0.5, color='lightpink', edgecolor='black')
plt.hist(x3, 10, rwidth=0.5, color='lightgreen', edgecolor='black')

plt.xlabel('IDH por Estado', fontsize=15)
plt.ylabel('Frequência Absoluta', fontsize=15)
plt.title('Nível de desigualdade no Brasil', fontsize=18)
plt.legend(['1991', '2000', '2010'], loc=1)

plt.show()
```

```python
resposta = f'''Item 2.
A partir do gráfico é possível notar que houve crescimento do IDH dos estados ao longo das três décadas.
'''
print(resposta)
```

```python
df_ano1991Filtrado = df[ df["ano"] == 1991]
df_ano2000Filtrado = df[ df["ano"] == 2000]
df_ano2010Filtrado = df[ df["ano"] == 2010]

x1 = df_ano1991Filtrado.expectativa_vida.values
x2 = df_ano2000Filtrado.expectativa_vida.values
x3 = df_ano2010Filtrado.expectativa_vida.values

y1 = df_ano1991Filtrado.idhm.values
y2 = df_ano2000Filtrado.idhm.values
y3 = df_ano2010Filtrado.idhm.values

plt.scatter(x1, y1, color='lightblue', edgecolor='black')
plt.scatter(x2, y2, color='lightpink', edgecolor='black')
plt.scatter(x3, y3, color='lightgreen', edgecolor='black')

plt.xlabel('Expectativa de Vida', fontsize=15)
plt.ylabel('IDH por Estado Brasileiro', fontsize=15)
plt.title('IDH x Expectativa de Vida', fontsize=18)
plt.legend(['1991', '2000', '2010'], loc=0)

plt.show()
```

```python
resposta = f'''Item 3.
A partir do gráfico é possível constatar duas observações: 
1ª) o crescimento do IDH brasileiro ao longo das três décadas e
2ª) para índices de IDH maiores, a expectativa de vida do brasileiro também cresce, independente da década.
'''
print(resposta)
```
